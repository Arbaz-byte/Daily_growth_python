# Dot Product

The dot product (or scalar product) multiplies two equal-length vectors to return a single number.

## Formulas

### Algebraic Definition
Multiplies corresponding components and adds the results.
$$a \cdot b = \sum_{i=1}^{n} a_i b_i = a_1b_1 + a_2b_2 + \dots + a_nb_n$$

### Geometric Definition
Multiplies vector magnitudes by the cosine of the angle $\theta$ between them.
$$a \cdot b = \|a\| \|b\| \cos(\theta)$$

## Key Interpretations

* **Zero (0):** Vectors are perpendicular ($90^\circ$ angle).
* **Positive:** Vectors point in a similar direction (acute angle).
* **Negative:** Vectors point in opposite directions (obtuse angle).

## Example

* **Vectors:** $a = [1, 2]$, $b = [4, 6]$
* **Calculation:** $(1 \times 4) + (2 \times 6)$
* **Result:** $4 + 12 = 16$


```python
'''
computing the dot product of two vectors using three different ways
'''
import numpy as np


#Using for loop

def dot_product(v1,v2):
    
    #edge cases:
    if len(v1)!=len(v2):
        return "Error:length of vectors must be same."
             
    dot_sum=0    
    for i in range(len(v1)):
        dot_sum+=v1[i]*v2[i]
        
    return dot_sum

t1,t2=[],[]
p1,p2=[5,6],[6]
v1,v2=[4,5,6],[8,9,6]
print("Using For loop:- ")
print("dot product is : ",dot_product(t1,t2))
print(dot_product(p1,p2))
print("dot product is : ",dot_product(v1,v2))


#using np.dot in numpy
print("Using np.dot :-")
print("dot product is :",np.dot(t1,t2))
print("dot product is :",np.dot(v1,v2))

#using @ operator
x=np.array([4,5,6])
y=np.array([8,9,6])

print("using @ operator :-")
print("dot product is :",(x@y))


           
```

# Euclidean Distance

Euclidean distance is the straight‑line distance between two points or vectors.

## Formula

For vectors **a** and **b** of length *n*:

$$\text{distance} = \sqrt{(a_1 - b_1)^2 + (a_2 - b_2)^2 + \dots + (a_n - b_n)^2}$$

## Example

* **Vectors:** $a = [1, 2]$, $b = [4, 6]$
* **Differences:** $[-3, -4]$
* **Squares:** $9 + 16 = 25$
* **Square root:** $\sqrt{25} = 5$


```python
def euclidean_dis(v1,v2):
    
    #edge case 
    if len(v1) != len(v2):
        return "Error : Length of vectors must be same."
    
    sqr_sum=0
    for i in range(len(v1)):
        sqr_sum+= np.square(v2[i]-v1[i])
              
    d= np.sqrt(sqr_sum)
    
    return d


#case:1
v1=np.array([1,2])
v2=np.array([4,6])
print("Distance : ",euclidean_dis(v1,v2))

#case:2
p1=np.array([])
p2=np.array([])
print("Distance : ",euclidean_dis(p1,p2))

#case:3
d1=np.array([5,6,5])
d2=np.array([5,6])
print(euclidean_dis(d1,d2))


```

# Cramer's Rule

Cramer's rule is an explicit formula for solving a system of linear equations with as many equations as unknowns, valid whenever the system has a unique solution.

## The Rule

For a system of linear equations represented in matrix form as $Ax = b$, where $A$ is an $n \times n$ coefficient matrix and $D = \det(A) \neq 0$:

The value of each variable $x_i$ is given by:
$$x_i = \frac{D_i}{D}$$

Where:
* **$D$** is the determinant of the coefficient matrix $A$.
* **$D_i$** is the determinant of the matrix formed by replacing the $i$-th column of $A$ with the constant column vector $b$.

## Matrix Breakdowns

### 2x2 System
Given the system:
$$a_1x + b_1y = c_1$$
$$a_2x + b_2y = c_2$$

* **$D$**: $\begin{vmatrix} a_1 & b_1 \\ a_2 & b_2 \end{vmatrix}$
* **$D_x$**: $\begin{vmatrix} c_1 & b_1 \\ c_2 & b_2 \end{vmatrix}$ (Main determinant with $x$-column replaced by constants)
* **$D_y$**: $\begin{vmatrix} a_1 & c_1 \\ a_2 & c_2 \end{vmatrix}$ (Main determinant with $y$-column replaced by constants)

Solutions:
$$x = \frac{D_x}{D}, \quad y = \frac{D_y}{D}$$

### 3x3 System
Given the system:
$$a_1x + b_1y + a_3z = d_1$$
$$a_2x + b_2y + b_3z = d_2$$
$$a_3x + b_3y + c_3z = d_3$$

* **$D$**: $\begin{vmatrix} a_1 & b_1 & c_1 \\ a_2 & b_2 & c_2 \\ a_3 & b_3 & c_3 \end{vmatrix}$
* **$D_x$**: $\begin{vmatrix} d_1 & b_1 & c_1 \\ d_2 & b_2 & c_2 \\ d_3 & b_3 & c_3 \end{vmatrix}$ (Main determinant with $x$-column replaced by constants)
* **$D_y$**: $\begin{vmatrix} a_1 & d_1 & c_1 \\ a_2 & d_2 & c_2 \\ a_3 & d_3 & c_3 \end{vmatrix}$ (Main determinant with $y$-column replaced by constants)
* **$D_z$**: $\begin{vmatrix} a_1 & b_1 & d_1 \\ a_2 & b_2 & d_2 \\ a_3 & b_3 & d_3 \end{vmatrix}$ (Main determinant with $z$-column replaced by constants)

Solutions:
$$x = \frac{D_x}{D}, \quad y = \frac{D_y}{D}, \quad z = \frac{D_z}{D}$$

## Conditions and Limitations
* **Square Systems Only:** The number of equations must equal the number of variables.
* **Unique Solution ($D \neq 0$):** If $D = 0$, the system has either no solution or infinitely many solutions, and Cramer's Rule cannot be used.
* **Efficiency:** It is highly inefficient for systems larger than 3x3 due to the high computational cost of calculating multiple large determinants.


```python
def cramers_compute(A,B):
    
    row,cols=A.shape
    if row != cols:
        return "Limitation Error: cramer's rule works for square matrices only."
    elif A.size==0 or B.size==0:
        return "Error : Matrix is Empty"
    
    sol=[]
    det=[]
    D = np.linalg.det(A)
    
    if np.isclose(D, 0):
        return "No Uniqe Solution(No solution or Infinitely many solutions))"
    
    
    for i in range(row):
        copy_A=A.copy()
        copy_A[:,i]=B.flatten() 
        det.append(round(np.linalg.det(copy_A)) )
        sol.append(det[i]/D )
        
    sol_array=np.array(sol)
    det_array=np.array(det)
    
    return sol_array,det_array 

a=np.array([ [2,3],
            [1,-1] ])

b = np.array([[8],
              [-1]]) 

solution,determinant = cramers_compute(a,b)
print("Solution for 2 Unkowns : ")
print (solution)



A = np.array([
    [2,  3, -1,  4,  2],
    [0,  1,  2,  5, -1],
    [0,  0,  3, -2,  6],
    [0,  0,  0,  2,  1],
    [0,  0,  0,  0,  2]
])

B = np.array([
    [4],
    [11],
    [3],
    [7],
    [6]
])

solution,determinant = cramers_compute(A,B)
print("Solution matrix for 5 unknows : ")
print (np.round(solution, 3))



#Using numpy
solution = np.linalg.solve(A, B)

print("Native NumPy Solution:")
print(np.round(solution.flatten(), 3))
           
        
        
        
    
    
```
